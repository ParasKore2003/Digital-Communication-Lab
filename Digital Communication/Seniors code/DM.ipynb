```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
amplitude = 5
frequency = 100  # Hz
duration = 1  # Second
sampling_rate = 10000  # Samples per second

# Time array
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate sinusoidal signal
signal = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the signal
plt.figure(figsize=(10, 4))
plt.plot(t, signal, color='blue')
plt.title('Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Adjust parameters
duration = 0.02  # Duration for two cycles
sampling_rate = 5 * 2 * frequency  # 5 times Nyquist rate

# Time array
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate sinusoidal signal
signal = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the signal
plt.figure(figsize=(10, 4))
plt.stem(t, signal, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.title('Sinusoidal Signal for Two Cycles')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python

# Staircase approximation step size
step_size = (2 * np.pi * amplitude * frequency) / sampling_rate
print(step_size)
# Staircase approximation
quantized_signal = np.round(signal / step_size) * step_size

# Plot the signals
plt.figure(figsize=(10, 6))

# Plot original sinusoidal signal
plt.plot(t,signal, label='Original Sinusoidal Signal', color='blue')

# Plot staircase approximation
plt.step(t, quantized_signal, label='Staircase Approximation', where='post', color='red')

plt.title('Staircase Approximation to Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
